```python
import pandas as pd
import numpy as np
#course = pd.read_csv("course.csv") 
#course_element = pd.read_csv("course_element.csv") 
course_module = pd.read_csv("course_module.csv") 
#sol_log = pd.read_csv("solution_log.csv") 
u_course_progress = pd.read_csv("user_course_progress.csv") 
#u_element_progress = pd.read_csv("user_element_progress.csv") 
u_module_progress = pd.read_csv("user_module_progress.csv")
```

```python
users = pd.read_csv("user_course_ids.csv")
u_course_progress = u_course_progress.merge(users, on=['user_id', 'course_id'])
data_time_begin = u_course_progress.time_created
data_time_end = u_course_progress.time_updated
u_course_progress.head(10)
```

```python

u_course_progress.time_created = pd.to_datetime(u_course_progress.time_created)
u_course_progress.time_updated = pd.to_datetime(u_course_progress.time_updated)
u_course_progress['time_lead'] = u_course_progress['time_updated'] - u_course_progress['time_created']
u_course_progress['time_lead_sec'] = u_course_progress['time_lead'].dt.total_seconds()
u_course_progress['speed'] = 1000000 * (u_course_progress['modules_achieved'] ) / u_course_progress['time_lead_sec']
u_course_progress['speed1'] = 1000000 *  u_course_progress['advanced_modules_done'] / u_course_progress['time_lead_sec']
u_course_progress['speed2'] = 1000000 * (u_course_progress['modules_achieved'] - u_course_progress['advanced_modules_done']) / u_course_progress['time_lead_sec']
#new = u_module_progress.loc[u_module_progress['user_id'] == 100120060110982009]
#new1 = course_module.loc[course_module['id'] == 52]
u_course_progress
```

```python
u_course_progress = u_course_progress.drop('advanced_modules_done', 1)
u_course_progress = u_course_progress.drop('time_created', 1)
u_course_progress = u_course_progress.drop('time_updated', 1)
u_course_progress = u_course_progress.drop('modules_closed', 1)
u_course_progress = u_course_progress.drop('modules_done', 1)
u_course_progress = u_course_progress.drop('time_lead', 1)
u_course_progress = u_course_progress.drop('time_lead_sec', 1)
u_course_progress
```

```python
u_course_progress = u_course_progress.drop('id', 1)
```

```python
u_course_progress.to_csv("speeds.csv", index = False)
```
